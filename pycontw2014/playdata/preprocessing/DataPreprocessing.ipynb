## Datetime Manipulation

````python
import datetime
import re
import numpy as np

def convert_timestr(xx):
    try:
        
        
        yy = re.search(ur"(:?Mon|Tue|Wed|Thu|Fri|Sat|Sun)\s*(:?Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)\s*([0-9]+)\s*([0-9+]+)\s*:\s*([0-9]{2})\s*:\s*([0-9]{2})\s*([0-9]{4})",
                       xx)
            
        if yy != None:
            temp = datetime.datetime.strptime(" ".join(yy.groups()),"%a %b %d %H %M %S %Y")
             
        else:
            yy = re.search(ur"(?P<m>[0-1]*[0-9])/(?P<d>[0-9]+)/(?P<y>[0-9]{4})\s*(?P<H>[0-9+]+)\s*:\s*(?P<M>[0-9]+)\s*:\s*(?P<S>[0-9]+)","01/26/2005 12:34:22 Wed").groupdict()
        
            temp = datetime.datetime(int(yy["y"]),
                                     int(yy["m"]),
                                     int(yy["d"]),
                                     int(yy["H"]),
                                     int(yy["M"]),
                                     int(yy["S"]))
#        print "[successfully] yy = ",yy
        return temp
    
    except Exception as e:
#    except:
        if np.isnan(xx):
            pass
#            print "xx == np.nan"
        else:
            print "Error = ",e
            print "[fail] xx = ",xx
            print  "Error Formate:%s" % xx
        return "Error Formate"


print articles_df["time"][1:1000].map(convert_timestr)
````

# ts_ngram ... 

````python
def ts_ngram(ts, n, name_prefix="x", remove_index = False):
    ngram_ts_list = [ts.shift(i) for i in range(n)]
    colnames = ["%s%i" % (name_prefix,i) for i in range(n)]
    
    if remove_index:
        return pd.concat(ngram_ts_list, axis=1,join="inner",keys=colnames)[n-1:].reset_index(drop=True)
    else:
        return pd.concat(ngram_ts_list, axis=1,join="inner",keys=colnames)[n-1:]
````

# Convert Data to Orange DataTable 

````python
import os 
import Orange

def convert_dataframe_to_orange(df, with_index=False, index_label=None):
    if with_index:
        df.to_csv('_temp_.csv',index = with_index, index_label=index_label)
    else:
        df.to_csv('_temp_.csv',index = with_index)
        
    orange_table = Orange.data.Table('_temp_.csv')
    os.unlink('_temp_.csv')
    return orange_table
````
